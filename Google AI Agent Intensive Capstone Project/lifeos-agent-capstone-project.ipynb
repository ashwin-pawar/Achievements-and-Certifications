```python
# ENVIRONMENT SETUP 

import json
import os

print("Environment ready (Offline Mode)")
```

```python
# MOCK GEMINI (Offline AI Simulation)

def mock_gemini(prompt: str) -> str:
    # You can modify this to match expected outputs
    return "Simulated AI Response for: " + prompt[:50] + "..."
```

```python
## AGENT 1 : ğŸ”µ Task Analyzer Agent (v1)

class TaskAnalyzerAgent:
    def analyze(self, user_input: str):
        
        # Mock call to AI
        response = mock_gemini("Extract tasks: " + user_input)

        # Simulated output (normally AI se aata)
        return [
            {"task": "Prepare slides"},
            {"task": "Pay electricity bill"},
            {"task": "Finish assignment"},
            {"task": "Go to gym"},
            {"task": "Call mom"}
        ]
```

```python
## AGENT 2 : ğŸŸ£ Priority Agent (v1)

class PriorityAgent:
    def prioritize(self, tasks):
        
        # Mock logic: assign simple priority levels
        prioritized = []
        for i, t in enumerate(tasks):
            t["priority"] = ["High", "Medium", "Low"][i % 3]
            t["estimated_time"] = "30 minutes"
            prioritized.append(t)
        return prioritized
```

```python
## AGENT 3: ğŸŸ  Daily Planner Agent (v1)

class DailyPlannerAgent:
    def create_plan(self, prioritized_tasks):
        
        # Mock logic: convert tasks into a time-based schedule
        schedule = []
        current_time = 9  # Start day at 9 AM

        for t in prioritized_tasks:
            schedule.append({
                "task": t["task"],
                "priority": t["priority"],
                "start_time": f"{current_time}:00",
                "end_time": f"{current_time+1}:00"
            })
            current_time += 1  # Each task = 1-hour block

        return schedule
```

```python
# AGENT 4: ğŸŸ¢ Summary Agent

class SummaryAgent:
    def generate_summary(self, day_plan):
        
        # Mock logic for summary creation
        total_tasks = len(day_plan)
        high_priority = sum(1 for t in day_plan if t["priority"] == "High")

        summary_text = (
            f"Total tasks planned: {total_tasks}. "
            f"High priority tasks: {high_priority}. "
            f"You are set for a productive day!"
        )

        return summary_text
```

```python
## ğŸ”§ Tools Used 
##TOOL 1: TIME CALCULATOR TOOL

def estimate_task_times(tasks):
    
    # Mock logic: every task takes 1 hour
    for t in tasks:
        t["estimated_time"] = "1 hour"
    return tasks
```

```python
## ğŸ’¾ Memory System

SESSION_MEMORY = {
    "past_plans": []
}
```

```python
# MAIN AGENT ğŸ”— Pipeline (v1)

# Initialize agents
analyzer = TaskAnalyzerAgent()
prioritizer = PriorityAgent()
planner = DailyPlannerAgent()
summarizer = SummaryAgent()

def life_os_pipeline(user_input: str):
    
    # Step 1: Analyze
    tasks = analyzer.analyze(user_input)

    # Step 2: Estimate times
    tasks = estimate_task_times(tasks)

    # Step 3: Prioritize
    prioritized = prioritizer.prioritize(tasks)

    # Step 4: Create plan
    plan = planner.create_plan(prioritized)

    # Step 5: Summary
    summary = summarizer.generate_summary(plan)

    # Step 6: Save to memory
    SESSION_MEMORY["past_plans"].append(plan)

    return {
        "tasks": tasks,
        "prioritized": prioritized,
        "plan": plan,
        "summary": summary
    }
```

## Day 3 â€” Full Logic Upgrade


```python
## ğŸ”µ Task Analyzer Agent (v2 â€” Improved)

class TaskAnalyzerAgentV2:
    def analyze(self, user_input: str):
        # Split by comma or 'and'
        parts = user_input.replace(" and ", ",").split(",")

        tasks = []
        for p in parts:
            clean = p.strip().lower()
            if clean:
                tasks.append({"task": clean})

        if not tasks:
            tasks.append({"task": "unspecified task"})

        return tasks
```

```python
## ğŸŸ£ Priority Agent (v2 â€” Improved)
class PriorityAgentV2:
    def prioritize(self, tasks):
        prioritized = []
        for t in tasks:
            txt = t["task"]

            if any(x in txt for x in ["urgent", "today", "asap", "tonight"]):
                t["priority"] = "High"
            elif any(x in txt for x in ["call", "gym", "clean"]):
                t["priority"] = "Medium"
            else:
                t["priority"] = "Low"

            t["estimated_time"] = "1 hour"
            prioritized.append(t)

        return prioritized
```

```python
## ğŸŸ  Daily Planner Agent (v2 â€” Improved)


class DailyPlannerAgentV2:
    def create_plan(self, prioritized_tasks):
        schedule = []
        current_hour = 9

        for t in prioritized_tasks:
            schedule.append({
                "task": t["task"],
                "priority": t["priority"],
                "start_time": f"{current_hour}:00",
                "end_time": f"{current_hour+1}:00",
            })

            current_hour += 1

            if current_hour == 13:
                schedule.append({
                    "task": "Lunch Break",
                    "priority": "-",
                    "start_time": "13:00",
                    "end_time": "14:00"
                })
                current_hour = 14

        return schedule
```

```python
# WEEKLY PLANNER AGENT (v2 â€” Improved)
class WeeklyPlannerAgent:
    def generate_weekly_plan(self, past_plans):
        if not past_plans:
            return "Not enough history for weekly plan."

        weekly = []
        task_count = {}

        for day in past_plans:
            for item in day:
                task = item["task"]
                task_count[task] = task_count.get(task, 0) + 1

        for task, count in task_count.items():
            weekly.append(f"{task} â†’ repeated {count} times this week")

        return weekly
```

```python
# SUMMARY AGENT (v2 â€” Improved)

class SummaryAgent:
    def generate_summary(self, day_plan):
        # Count tasks
        total_tasks = len(day_plan)
        high_priority = sum(1 for t in day_plan if t["priority"] == "High")

        summary_text = (
            f"Total tasks planned: {total_tasks}. "
            f"High priority tasks: {high_priority}. "
            f"You are set for a productive day!"
        )

        return summary_text
```

```python
## ğŸ”— PIPELINE (V2)

analyzer2 = TaskAnalyzerAgentV2()
prioritizer2 = PriorityAgentV2()
planner2 = DailyPlannerAgentV2()
summarizer = SummaryAgent()
weekly_agent = WeeklyPlannerAgent()

def life_os_pipeline_v2(user_input):
    tasks = analyzer2.analyze(user_input)
    prioritized = prioritizer2.prioritize(tasks)
    plan = planner2.create_plan(prioritized)
    summary = summarizer.generate_summary(plan)
    
    # Save plan for weekly review
    SESSION_MEMORY["past_plans"].append(plan)

    weekly_plan = weekly_agent.generate_weekly_plan(SESSION_MEMORY["past_plans"])

    return {
        "tasks": tasks,
        "prioritized": prioritized,
        "plan": plan,
        "summary": summary,
        "weekly_plan": weekly_plan
    }
```
